# This page describes the splitting and stacking of matrices using NumPy inbuilt functions.

```python
import os.path
from skimage.io import imread
from skimage import data_dir
import matplotlib.pyplot as plt
import numpy as np
# Original Image
img = imread(os.path.join(data_dir, 'ihc.png'))
plt.imshow(img)
```

### 1. Splitting a matrix

**We split the given image matrix into 2 blocks horizontally and 4 blocks vertically.**

```python
h_slice = img.copy()
h_slice = np.hsplit(h_slice,2)
''' Horizontal Split '''
plt.figure()
plt.imshow(h_slice[1])
```

```python
v_slice = img.copy()
v_slice = np.vsplit(v_slice,4)
''' Vertical Split '''
plt.figure()
plt.imshow(v_slice[1])
```

### 2. Merging matrices

**To merge different matrices (on either axis), NumPy provides two separate functions namely vstack and hstack.**

```python
h_stack = np.hstack( (h_slice[1],h_slice[0]) )
''' Horizontal Merge '''
plt.figure()
plt.imshow(h_stack)
```

```python
v_stack = np.vstack( (v_slice[1],v_slice[0]) )
''' Vertical Merge '''
plt.figure()
plt.imshow(v_stack)
```
