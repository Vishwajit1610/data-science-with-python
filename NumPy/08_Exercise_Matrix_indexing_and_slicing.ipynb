# Problem Statement:

## Open the Phantom image available inbuilt under scikit-image module and do the following:

#### 1. Upload the image into python environment and show the image. 

#### 2. Perform black and white thresholding with decision boundary set at 0.15. Store the new image in a variable new_img and show it.

#### 3. Use the numpy array manipulating function fliplr() to get the mirror image.

#### 4. Compress the flipped image along both the axes by 50% in sequence. 

```python
# Importing Modules
import numpy as np
import matplotlib.pyplot as plt 
import skimage.data
```

**1. Upload the image and display**

```python
img = skimage.data.shepp_logan_phantom()
```

```python
plt.imshow(img)
```

**2. Black and White thresholding**

```python
threshold_val = 0.15
new_img = (img > threshold_val).astype(np.uint8) 
```

```python
print("Thresholded Image:")
plt.imshow(new_img)
```

**3. Mirror Image using fliplr()**

```python
mirror_img = np.fliplr(new_img)
```

```python
plt.imshow(mirror_img)
```

**4. Compressing Flipped image along both the axes by 50% in sequence**

```python
compressed_img = mirror_img[: : 2,: : 2]
```

```python
plt.imshow(compressed_img)
```
