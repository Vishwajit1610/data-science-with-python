# Plot Texts

## Problem Statement:

#### Given values for x and y, arrive at the following plot shown below:
```         
import numpy as np
x = np.arange(0.0, 5.0, 0.1)
y = np.cos(2 * np.pi * x1) * np.exp(-x1)
```

### Expected Output:
![A Graph](img/Exercise_Plot_Texts_Output.png)

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
x = np.arange(0.0, 5.0, 0.1)
y = np.cos(2 * np.pi * x) * np.exp(-x)
````

**Plot Graph**

````python
fid, ax = plt.subplots(figsize = (8, 6))

ax.plot(x, y, 'k-')

# Title
plt.suptitle('Main heading', fontsize=16, fontweight='bold')
plt.title('Figure Title')

# Labels
ax.set_xlabel('Abscissa', color='green')
ax.set_ylabel('Ordinate', color='green')

# Setting Grid
ax.grid(True)

# Annotations

# Annotations 1
peak1_y = np.exp(-1)
ax.text(1, peak1_y + 0.03, # Position text slightly above the peak
        r'$\theta = 60^\circ$', # Raw string for LaTeX
        horizontalalignment='center') # Center the text on the x-coordinate

# Annotations 2
peak2_y = np.exp(-2)
ax.annotate('2nd crest',
            xy=(2.0, peak2_y),             # (x,y) point to arrow *to*
            xytext=(2.5, 0.3),             # (x,y) location of the text
            arrowprops=dict(arrowstyle='->', color='green', lw=2.5)
           )

plt.show()
````
