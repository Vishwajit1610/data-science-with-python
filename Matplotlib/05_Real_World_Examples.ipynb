### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Real World Examples

## Line Graph

#### We will use the Fifa CSV Dataset

**The file should be present in this directory for this code to work.**

```python
fifa = pd.read_csv('fifa_data.csv')
```

```python
plt.figure(figsize=(5, 3), dpi = 150)

plt.title('Fifa Players Names and Ratings', fontdict = {'family': 'Times New Roman', 'fontsize': 15, 'fontweight': 'bold'})

plt.plot(fifa.Name.head(10), fifa.Overall.head(10), 'r.--', label = 'Rating')

plt.xticks(rotation = 90, ha = 'center')

plt.xlabel('Player Names')
plt.ylabel('Player Rating')

plt.legend()

plt.show()
```

## Histogram

```python
bins = [40, 50, 60, 70,  80, 90, 100]

plt.hist(fifa.Overall, bins = bins, color = 'r')

plt.xticks(bins)

plt.ylabel('Number of Players')
plt.xlabel('Rating of Players')
plt.title('Distribution Graph of Players Skills in FIFA 2018')

plt.show()
```

## Pie Chart

```python
left = (fifa['Preferred Foot'] == 'Left').sum()
right = (fifa['Preferred Foot'] == 'Right').sum()
```

```python
labels = ['Left', 'Right']

colors = ['y', 'r']

plt.pie([left, right], labels = labels, colors = colors, autopct = '%.2f %%')

plt.title('Foot Preference of FIFA Players')

plt.show()
```
