## Scatter PLot

### Now let us try to figure out if there is any relation between the Pokemon catch rate and total power. 

### To do so, we plot two features namely Total and Catch_Rate against each other using scatter plot.

**Import Necessary Modules**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('pokemon_alopez247.csv')
```

```python
tot_power = df.iloc[:,4]
print(tot_power.head(4))
```

```python
catch_rate = df.iloc[:, 21]
print(catch_rate.head(4))
```

**Plot Scatter Graph**

```python
fig, ax = plt.subplots()

p = ax.scatter(catch_rate, tot_power, color = 'g')

ax.grid()

ax.set_xlabel('Catch Rate')
ax.set_ylabel('Total Power')
ax.set_title('Pokemon Catch Rate vs their Power')

plt.legend([p],['Pokemons'])

plt.show()
```

### We can observe that there are various Pokemon who despite having less Power still have less Catch Rate. 

### For more clarity, we can highlight one such common group present at catch rate 45. 

### Matplotlib provides patches through which we can make an overwriting to highlight specified area on our previous plot.

```python
import matplotlib.patches as patches
import matplotlib.transforms as transforms

trans = transforms.blended_transform_factory(
    ax.transData,ax.transAxes)

rect = patches.Rectangle((44,0), width=2, height=5,
                         transform=trans, color='red',
                         alpha=0.4)
ax.add_patch(rect)

fig
```

### Since we have visualized that there are some adamant Pokemon who have fairly less catch rate despite having less power. 

### So, let us name few of them with conditions Catch_Rate equals 45 and Power less than or equal to 330.

```python
catch_rate_45 = df[df.loc[:,'Catch_Rate'] == 45]

pow_330 = catch_rate_45[catch_rate_45.loc[:,'Total'] <= 330]

print("Number of such Pokemons:", len(pow_330))

# Top 10 adamant Pokets
print(pow_330.loc[:,'Name'].head(10))
```
