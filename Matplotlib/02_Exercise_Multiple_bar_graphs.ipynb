## Problem Statement:
### Given score points of Men and Women in a certain game played 5 times with the corresponding errors in the observations:

```
n_grps = np.arange(5)
bar_width = 0.4
men_sc = [20, 30, 10, 50, 90]
err_men_sc = [2, 3, 4, 5, 4]
women = [10, 123, 19, 60, 40]
err_women_sc = [1, 6, 2, 8, 7]
```

### Plot a bar graph depicting scores for all 5 games along with errors in observation. Take colors of your choice.

````python
# Importing Modules
import numpy as np
import matplotlib.pyplot as plt
````

````python
# Data
n_grps = np.arange(5)
bar_width = 0.4
men_sc = [20, 30, 10, 50, 90]
err_men_sc = [2, 3, 4, 5, 4]
women_sc = [10, 123, 19, 60, 40]
err_women_sc = [1, 6, 2, 8, 7]
````

````python
# Setup Figure and Positions
fig, ax = plt.subplots(figsize=(8, 6))
position_men = n_grps - bar_width / 2
position_women = n_grps + bar_width / 2

# Plot Bars
ax.bar(position_men, men_sc, bar_width, 
       yerr=err_men_sc, 
       label='Men', 
       color='red', 
       alpha=0.6, 
       capsize=5)

ax.bar(position_women, women_sc, bar_width, 
       yerr=err_women_sc, 
       label='Women', 
       color='pink', 
       alpha=0.6, 
       capsize=5)

# Finalize Plot
ax.set_ylabel('Scores')
ax.set_title('Scores by Men and Women in 5 Games')
ax.set_xticks(n_grps)
ax.set_xticklabels(('A', 'B', 'C', 'D', 'E'))

# Add Legend
ax.legend()

# Show the Plot
plt.show()
````
